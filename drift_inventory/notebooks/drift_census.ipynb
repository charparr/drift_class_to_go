```python
def find_drift_thresholds(ds):

    for k in ds.keys():
    
        # Thresholds to test (percentages of the mean depth)
        thresholds = np.round(np.arange(0.8, 2.3, 0.1) * ds[k]['mu'], 2)
        
        # Computing total area / volume of all snow pixels
        pxl_sz = 1
        total_area = (~np.isnan(ds[k]['arr'])).sum() * pxl_sz
        total_volume = np.nansum((~np.isnan(ds[k]['arr'])) * ds[k]['arr'] * pxl_sz)
        ds[k]['total area'] = total_area
        ds[k]['total volume'] = total_volume
        
#         print(total_area)
#         print(total_volume)

        
    
#     # Compute depth thresholds
        d = dict()
    
        for i in thresholds:
            
            t = str(i) + ' m'
            
            drift_mask = (ds[k]['arr'] >= i)
            depth_drift_masked = drift_mask * ds[k]['arr']
            depth_drift_masked[depth_drift_masked == 0] = np.nan

            not_drift_mask = (ds[k]['arr'] < i)
            depth_not_drift_masked = not_drift_mask * ds[k]['arr']
            depth_not_drift_masked[depth_not_drift_masked == 0] = np.nan

            d[t] = {}
            d[t]['drift_area'] = int(np.nansum(drift_mask))
            d[t]['not_drift_area'] = int(np.nansum(not_drift_mask))
            d[t]['drift_volume'] = int(np.nansum(drift_mask * ds[k]['arr'] * pxl_sz))
            d[t]['not_drift_volume'] = int(np.nansum(not_drift_mask * ds[k]['arr'] * pxl_sz))

        
        # Move to dict to df
        df = pd.DataFrame.from_dict(d).T
        df['Drift Threshold (pct. of mean depth)'] = np.arange(0.8, 2.3, 0.1) * 100
        df['Drift Area pct.'] = df.drift_area / total_area * 100
        df['Not Drift Area pct.'] = df.not_drift_area / total_area * 100
        df['Drift Volume pct.'] = df.drift_volume / total_volume * 100
        df['Not Drift Volume pct.'] = df.not_drift_volume / total_volume * 100
    
        # Find the inflection Threshold
        df['Drift Volume-Area Difference (pct.)'] = df['Drift Volume pct.'] - df['Drift Area pct.']
        df['Drift Volume-Area Difference Slope'] = np.gradient(df['Drift Volume-Area Difference (pct.)'])
        df.set_index(df['Drift Threshold (pct. of mean depth)'], inplace=True)
        ds[k]['pct of mean threshold'] = df['Drift Volume-Area Difference Slope'].idxmin()
        ds[k]['depth threshold'] = (ds[k]['pct of mean threshold'] / 100.0) * ds[k]['mu']
        ds[k]['threshold test results'] = df
```

```python
find_drift_thresholds(d)
```

```python
for k in d.keys():
    df = d[k]['threshold test results']
    t = d[k]['pct of mean threshold']
    d[k]['Drift Area %'] = df.loc[t]['Drift Area pct.']
    d[k]['Drift Volume %'] = df.loc[t]['Drift Volume pct.']
    d[k]['Drift Area'] = df.loc[t]['drift_area']
    d[k]['Drift Volume'] = df.loc[t]['drift_volume']
    d[k]['drift mask'] = d[k]['arr'] >= d[k]['depth threshold']
```

```python
fldarea = [k.split('/')[2] for k in d.keys()] 
year = [d[k]['year'][0] for k in d.keys()]
dareas = [d[k]['Drift Area %'] for k in d.keys()]
dvols = [d[k]['Drift Volume %'] for k in d.keys()]
dareamag = [d[k]['Drift Area'] / 1000000 for k in d.keys()]
dvolmag = [d[k]['Drift Volume'] for k in d.keys()]
pctthresh = [d[k]['pct of mean threshold'] for k in d.keys()]
dthresh = [d[k]['depth threshold'] for k in d.keys()]
mus = [d[k]['mu'] for k in d.keys()]
sigs = [d[k]['sigma'] for k in d.keys()]
cvs = [d[k]['CV'] for k in d.keys()]
tareas = [d[k]['total area'] / 1000000 for k in d.keys()]

tuples = list(zip(fldarea, year))
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['Field Area', 'Year'])
df = pd.DataFrame(index=index)
df['Field Area'] = fldarea
df['Year'] = year
df['Drift Area [% total]'] = dareas
df['Drift Volume [% total]'] = dvols
df['Drift Area [km^2]'] = dareamag
df['Drift Volume [m^3]'] = dvolmag
df['Drift Threshold [% of mean depth]'] = pctthresh
df['Drift Threshold [m]'] = dthresh
df['Mean Depth [m]'] = mus
df['SD Depth [m]'] = sigs
df['CV Depth'] = cvs
df['Total Area [km^2]'] = tareas

df['Area Normalized Drift Volume [m^3/km^2]'] = df['Drift Volume [m^3]'] / df['Total Area [km^2]']

df = df.round(2)
df
```

```python
sd['../depth_dDEMs/clpx/corrected/s_endclpx_depth_098_2015.tif']['threshold'] = 0.60

sd['../depth_dDEMs/clpx/corrected/s_endclpx_depth_096_2016.tif']['threshold'] = 0.46

sd['../depth_dDEMs/clpx/corrected/s_endclpx_depth_101_2017.tif']['threshold'] = 0.50

sd['../depth_dDEMs/clpx/corrected/s_endclpx_depth_105_2018.tif']['threshold'] = 0.82

for k in sd.keys():
    
    pxl_sz = 1
    drift_mask = sd[k]['arr'] >= sd[k]['threshold']
    total_area = (~np.isnan(sd[k]['arr'])).sum() * pxl_sz
    
    
    sd[k]['total area'] = total_area
    
    sd[k]['Drift Area'] = int(np.nansum(drift_mask))
    
    sd[k]['Drift Volume'] = int(np.nansum(drift_mask * sd[k]['arr'] * pxl_sz))

sfldarea = [k.split('/')[2] for k in sd.keys()] 
syear = [sd[k]['year'][0] for k in sd.keys()]
sdareamag = [sd[k]['Drift Area'] / 1000000 for k in sd.keys()]
sdvolmag = [sd[k]['Drift Volume'] for k in sd.keys()]

smus = [sd[k]['mu'] for k in sd.keys()]
ssigs = [sd[k]['sigma'] for k in sd.keys()]
scvs = [sd[k]['CV'] for k in sd.keys()]
stareas = [sd[k]['total area'] / 1000000 for k in sd.keys()]

stuples = list(zip(sfldarea, syear))

index = pd.MultiIndex.from_tuples(stuples, names=['Field Area', 'Year'])
sdf = pd.DataFrame(index=index)
sdf['Field Area'] = sfldarea
sdf['Year'] = syear
sdf['Drift Area [km^2]'] = sdareamag
sdf['Drift Volume [m^3]'] = sdvolmag
sdf['Mean Depth [m]'] = smus
sdf['SD Depth [m]'] = ssigs
sdf['CV Depth'] = scvs
sdf['Total Area [km^2]'] = stareas

sdf['Area Normalized Drift Volume [m^3/km^2]'] = sdf['Drift Volume [m^3]'] / sdf['Total Area [km^2]']

sdf = sdf.round(2)
sdf.sort_index(level='Year', inplace=True)
sdf
```
