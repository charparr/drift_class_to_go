```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('paper')
```

```python
df = pd.read_csv('../../figs/fig_tabular_data/aggregate_validation_results.csv')

df.head()
```

```python
plt.figure(figsize=(16,10))
plt.title('Validation Results Using ' + str(len(df)) + " MagnaProbe Field Measurements of Snow Depth")
plt.ylim(-1.1, 1.1)
sns.boxplot(x="Year", y="Probe-Raster Delta [m]", hue="Study Area", data=df, fliersize=2)
plt.ylabel('Probe Minus Snow Depth Map Difference [m]')

plt.savefig('../../figs/pngs/f5_validation_box_and_whisker_by_year_and_field_area.png',
            dpi=300, bbox_inches='tight')
```
